## Get reviews from link

```python
import requests
from bs4 import BeautifulSoup

def parse_webpage(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        print(soup)
        links = soup.span
        print(links)
    else:
        print("Failed to retrieve webpage")

# Example usage:
url = "https://www.google.com/maps/place/Freshroll/@37.8681077,-122.2593318,17z/data=!4m8!3m7!1s0x80857d5fd2e32abd:0x58e33166de67a490!8m2!3d37.8681077!4d-122.2593318!9m1!1b1!16s%2Fg%2F11j58gqwvc?entry=ttu"
parse_webpage(url)
```
